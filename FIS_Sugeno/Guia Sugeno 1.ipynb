```python
from importlib import reload  # Esto es necesario en algunos casos para recargar módulos

import numpy as np
from matplotlib import pyplot as plt
import sugeno
reload(sugeno)

#Leer data de txt
_input = []
with open("data_diodo.txt", 'r') as file:
    for line in file:
        x, y = map(float, line.strip().split())
        _input.append([x, y])
data = np.array(_input)

def graficar(dat):
    # Extraer las coordenadas x e y por separado
    x_coords = dat[:, 0]
    y_coords = dat[:, 1]
    # Graficar los puntos
    plt.scatter(x_coords, y_coords, color='blue', marker='o', s=7)
    # Etiquetas y título
    plt.xlabel('Tension')
    plt.ylabel('Corriente')
    plt.title('Datos Diodo')
    # Mostrar la gráfica
    plt.show()

graficar(data)

```

```python
data_x = data[:, 0]
data_y = data[:, 1]

fis2 = sugeno.fis()
fis2.genfis(data, 1.1)
fis2.viewInputs()
r = fis2.evalfis(np.vstack(data_x))

plt.figure()
plt.plot(data_x,data_y)
plt.plot(data_x,r,linestyle='--')

fis2.solutions
fis2.rules
```

```python
#
labels, centers = sugeno.subclust2(data, 0.5, 1.1)

labels
centers

# Extraer las coordenadas x e y por separado
x_coords_data = data[:, 0]
y_coords_data = data[:, 1]
x_coords = centers[:, 0]
y_coords = centers[:, 1]
# Graficar los puntos
plt.scatter(x_coords_data, y_coords_data, color='blue', marker='o', s=7)
plt.scatter(x_coords, y_coords, color='red', marker='o', s=20)
# Etiquetas y título
plt.xlabel('Tension')
plt.ylabel('Corriente')
plt.title('Datos Diodo')
# Mostrar la gráfica
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = 2
kmeans = KMeans(n_clusters = n_clusters, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
kmeans.fit(data)

labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

print(labels)
print(cluster_centers)

# Extraer las coordenadas x e y por separado
x_coords_data = data[:, 0]
y_coords_data = data[:, 1]
x_coords = cluster_centers[:, 0]
y_coords = cluster_centers[:, 1]
# Graficar los puntos
plt.scatter(x_coords_data, y_coords_data, color='blue', marker='o', s=7)
plt.scatter(x_coords, y_coords, color='red', marker='o', s=20)
# Etiquetas y título
plt.xlabel('Tension')
plt.ylabel('Corriente')
plt.title('Datos Diodo')
# Mostrar la gráfica
plt.show()
```
