# K-Means propio

```python
# Imports
import numpy as np
from matplotlib import pyplot as plt

def graficar(dat):
    # Extraer las coordenadas x e y por separado
    x_coords = dat[:, 0]
    y_coords = dat[:, 1]
    # Graficar los puntos
    plt.scatter(x_coords, y_coords, color='blue', marker='o', s=7)
    # Etiquetas y título
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Datos Iniciales')
    # Mostrar la gráfica
    plt.show()
    
def graficar2(dat, c):
    # Extraer las coordenadas x e y por separado
    x_coords = dat[:, 0]
    y_coords = dat[:, 1]
    x_coords2 = c[:, 0]
    y_coords2 = c[:,1]
    # Graficar los puntos
    plt.scatter(x_coords, y_coords, color='blue', marker='o', s=7, label='Datos')
    plt.scatter(x_coords2, y_coords2, color='red', marker='x', s=50, label='Centros')
    # Etiquetas y título
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Datos')
    # Mostrar la gráfica
    plt.show()


# Set three centers, the model should predict similar results
center_1 = np.array([1,1])
center_2 = np.array([5,5])
center_3 = np.array([8,1])

# Generate random data and center it to the three centers
data_1 = np.random.randn(50, 2) + center_1
data_2 = np.random.randn(50,2) + center_2
data_3 = np.random.randn(50,2) + center_3

data = np.concatenate((data_1, data_2, data_3), axis = 0)

graficar(data)
```

```python
#Centros iniciales
c = np.array(([-2,4],[10,0],[6,8]))
#c = np.array(([0,10],[10,10],[4,0]))

graficar2(data,c)
```

```python
#Cantidad de clusters K
k = c.shape[0]
#Cantidad de datos
n = data.shape[0]
#Array de pertenencia
p = np.full(n, -1)

def graficar_resultado():
    colores = ['red', 'blue', 'green']
    # Asignar colores a los puntos según su pertenencia
    colores_puntos = [colores[grupo] for grupo in p]
    # Graficar puntos
    plt.scatter(data[:, 0], data[:, 1], color=colores_puntos, marker='o', s=7, label='Datos')
    plt.scatter(c[:,0], c[:,1], color='black', marker='x', s=50, label='Centros')
    # Etiquetas y título
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')
    plt.title('Datos')
    # Mostrar la gráfica
    plt.show()

def calcular_pertenencia():
    for j in range(n):
        distancias = np.linalg.norm(data[j] - c, axis=1)
        #índice de la distancia mínima
        p[j] = np.argmin(distancias)

def calcular_centros():
    for i in range(k):
        c[i] = np.mean(data[p==i], axis=0)
    #print(c)
            
#Datos y centros de cluster iniciales
def k_means(data, c):
    calcular_pertenencia()
    graficar_resultado()
    for i in range(50):    
        calcular_centros()
        calcular_pertenencia()
    graficar_resultado()


k_means(data,c)

    
    
```
